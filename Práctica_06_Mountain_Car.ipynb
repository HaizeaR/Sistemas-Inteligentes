Definición de Hiper-Parámetros usando formularios

Los hiper-parámetros son parámetros ajustables que permiten controlar el proceso de entrenamiento del modelo y ajustarlos peude traducirse a mejores resultados.  

````python
#@title Hiper-parámetros

total_episodes = 10000       # @param {type:"integer"}
learning_rate = 0.2          # @param {type:"number"}
max_steps =                1000# @param {type:"integer"}
gamma = 0.9                 # @param {type:"number"}
````

````python
#@title Parámetros de exploración
epsilon =                  1.0# @param {type:"number"}
max_epsilon = 1.0             # @param {type:"number"}
min_epsilon = 0.01            # @param {type:"number"}
decay_rate = 0.005            # @param {type:"number"}
````

En base a las acciones que hemos tomado, el carro se mueve aplicando las leyes de la física. Cada vez que llamamos a la función ```env.step()```, nos devuelve una tupla con los siguientes valores:
- Un array con el nuevo estado (o las observaciones).
- La recompensa (un valor escalar de tipo ```float```)
- Un flag que determina si hemos legado al final o no.
- Un diccionario de Python con información adicional.

El objeto ```env``` tiene también un método ```render``` que permite visualizar enl entorno y los movimientos tanto del carro como de la polea.

Función esencial que se implementará en el código


```
Q’(s1, s2, a) = (1 — w)*Q(s1, s2, a) + w*(r+d*Q(s1’, s2’, argmax a’ Q(s1’, s2’, a’)))

```

Los pasos básicos para resolver el problema son los siguientes: 

1. Crear e inicializar la QTable
2. Mirar lo que ocurre en el estado actual - state = env.render
3. Decidir que acción tomar con [a] (if/else)
4. Realizar accion y observar el resultado obtenido de la misma (env.step(action))
5. Actualizar la QTable usando la ecuación anterior 
6. Repetir proceso 

````python
if COLAB:
  env = wrap_env (gym.make("MountainCar-v0"))
else:
  env = gym.make("MountainCar−v0")

# se define un array para guardar la puntuación
rewards = []

# 2 Por cada vida o fin de aprendizaje
for episode in range(total_episodes):
    # Reiniciamos le entorno
    state = env.reset()    
    step = 0 
    done = False
    total_rewards = 0
        
    #Se ajusta la variable estado
    state_ajustado = (state - env.observation_space.low)*np.array([10, 100])
    state_ajustado = np.round(state_ajustado, 0).astype(int)

    while done != True:

        if i >= (total_episodes - 20):
            env.render()

      #Calcular la próxima acción
        if np.random.random() < 1 - epsilon:
            action = np.argmax(qtable[state_ajustado[0], state_ajustado[1]])
            
        else:
            action = np.random.randint(0, n_acciones)
                
            
        # Llevamo a cabo la acción (a) y observamos el nuevo entorno (s') y la recompensa (r)
        new_state, reward, done, info = env.step(action)

        new_state_ajustado = (new_state - env.observation_space.low)*np.array([10, 100])
        new_state_ajustado = np.round(new_state_ajustado, 0).astype(int)

        #Allow for terminal states
        if done and new_state[0] >= 0.5:
            qtable[state_ajustado[0], state_ajustado[1], action] = reward
            
        # Adjust Q value for current state
        else:

          qtable[new_state_ajustado[0], new_state_ajustado[1], action] = qtable[new_state_ajustado[0], new_state_ajustado[1], action] + learning_rate * (reward + gamma * np.max(qtable[new_state_ajustado[0], new_state_ajustado[1]]) - qtable[new_state_ajustado[0], new_state_ajustado[1], action])
        
        total_rewards += reward
        # Acutalizamos nuestro estado
        state_ajustado = new_state_ajustado
        
        # Reducimos el valor de epsilon (ya que necesitamos menos exploration)
        epsilon = min_epsilon + (max_epsilon - min_epsilon)*np.exp(-decay_rate*episode) 
        rewards.append(total_rewards)
      
    print(f'Episodio {episode} ; Recompensa: {total_rewards}; Acción:{action}')

        

print ("Puntuación en el tiempo: " +  str(sum(rewards)/total_episodes))
env.close()
````
